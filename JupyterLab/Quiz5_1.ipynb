```python
import cv2
import numpy as np
```

```python
source = './Quiz5_1.jpg'
```

```python
img = cv2.imread(source, cv2.IMREAD_COLOR)
imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret1, thresh1 = cv2.threshold(imgGray, 127, 255, cv2.THRESH_BINARY)
contours1, hierarchy1 = cv2.findContours( thresh1, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

ret2, thresh2 = cv2.threshold(imgGray, 127, 255, cv2.THRESH_BINARY_INV)
contours2, hierarchy2 = cv2.findContours( thresh2, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

print('Number of contours case 1 : ', len(contours1))
print('Number of contours case 2 : ', len(contours2))
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the original image
source = 'Quiz5_3_2.png'
img = cv2.imread(source, cv2.IMREAD_COLOR)

# Apply Gaussian blur
blurred = cv2.GaussianBlur(img, (5, 5), 0)

# Apply erosion
kernel = np.ones((5, 5), np.uint8)
eroded = cv2.erode(blurred, kernel, iterations=1)

# Convert to grayscale
gray_blurred = cv2.cvtColor(blurred, cv2.COLOR_BGR2GRAY)
gray_eroded = cv2.cvtColor(eroded, cv2.COLOR_BGR2GRAY)

# Apply threshold to get a binary image
_, binary_blurred = cv2.threshold(gray_blurred, 127, 255, cv2.THRESH_BINARY)
_, binary_eroded = cv2.threshold(gray_eroded, 127, 255, cv2.THRESH_BINARY)

# Find contours
contours_blurred, _ = cv2.findContours(binary_blurred, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours_eroded, _ = cv2.findContours(binary_eroded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the images
blurred_with_contours = cv2.drawContours(blurred.copy(), contours_blurred, -1, (0, 255, 0), 2)
eroded_with_contours = cv2.drawContours(eroded.copy(), contours_eroded, -1, (0, 255, 0), 2)

# Save the images with contours
cv2.imwrite('Quiz5_3_2_blur_with_contours.png', blurred_with_contours)
cv2.imwrite('Quiz5_3_2_eroded_with_contours.png', eroded_with_contours)

# Display the original, blurred, and eroded images with contours
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Blurred Image with Contours')
plt.imshow(cv2.cvtColor(blurred_with_contours, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Eroded Image with Contours')
plt.imshow(cv2.cvtColor(eroded_with_contours, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

# Print the number of contours detected
print(f'Number of contours in blurred image: {len(contours_blurred)}')
print(f'Number of contours in eroded image: {len(contours_eroded)}')
```
